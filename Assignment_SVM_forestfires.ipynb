```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("C:/Users/admin/Desktop/Assignment_DataScience/forestfires.csv")
df
```

```python
df.describe()
```

```python
##Dropping the month and day columns
df.drop(["month","day"],axis=1,inplace =True)
```

```python
df
```

```python
X = df.iloc[:,0:28]
X
```

```python
Y = df.iloc[:,28]
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.25)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.svm import SVC
```

```python
model_linear = SVC(kernel = "linear")
model_linear.fit(x_train,y_train)
```

```python
pred_test_linear = model_linear.predict(x_test)
```

```python
np.mean(pred_test_linear==y_test) # Accuracy = 98.46%
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
acc = accuracy_score(y_test, pred_test_linear) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, pred_test_linear)
```

```python
# Kernel = poly
model_poly = SVC(kernel = "poly")
model_poly.fit(x_train,y_train)
```

```python
pred_test_poly = model_poly.predict(x_test)
```

```python
np.mean(pred_test_poly==y_test)
```

```python
# kernel = rbf
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(x_train,y_train)
```

```python
pred_test_rbf = model_rbf.predict(x_test)
```

```python
np.mean(pred_test_rbf==y_test) #Accuracy = 76.15%
```
